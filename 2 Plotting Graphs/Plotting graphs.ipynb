```python
import numpy as np
# x=np.array([1,2,3,4])
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,color="black",linewidth=5,label="x^3")
plt.plot(x,y2,color="red",linewidth=5,label="x^2")

# plt.axis([0,10,0,100])

plt.grid()
plt.text(2,80,"Text",fontsize=12)

plt.ylabel("Y")
plt.xlabel("X")
plt.legend()
plt.title("Matplotlib demo")

plt.show()
```

```python
#pie graphs
labels=["A","B","C","D"]
sizes=[3,4,6,2]
explode=[0.1,0,0,0]
colors=["blue","purple","red","pink"]
plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct="%.2f%%",counterclock=False,startangle=100)
plt.axis("equal")
plt.title("Split among classes")
plt.show()
```
