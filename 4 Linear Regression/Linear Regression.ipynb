```python
#Linear regression is basically assuming linear relation between x and y

#It tries to get best possible line or the best possible answer for given set of data
#let consider y=mx+c    It will try to predict one of the best value of m and c for any x and y
```

```python
import numpy as np
data=np.loadtxt("data.csv",delimiter=",")

```

```python
x=data[:,0].reshape(-1,1)
y=data[:,1]

x.shape
```

```python
from sklearn import model_selection

X_train,X_test,Y_train,Y_test=model_selection.train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression

alg1=LinearRegression()
alg1.fit(X_train,Y_train)
```

```python
alg1.coef_
#this is basically m
```

```python
alg1.intercept_
#this is c
```

```python
import matplotlib.pyplot as plt

m=alg1.coef_[0]
c=alg1.intercept_

x_line=np.arange(30,70,0.1)
y_line=m*x_line+c

train_1d=X_train.reshape(75)
plt.scatter(X_train,Y_train)
plt.plot(x_line,y_line,"r")
plt.show()
```
